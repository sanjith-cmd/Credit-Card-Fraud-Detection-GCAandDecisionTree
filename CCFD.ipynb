```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CSV_EXCEL_files/creditcard.csv')
print(df.head(10))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df['Amount']=sc.fit_transform(df['Amount'].values.reshape(-1,1))
df.head(10)
```

```python
df.drop(['Time'],axis=1,inplace=True)
df.head(10)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
fig=plt.figure(figsize=(20,70))
plt.suptitle('Histogram of Numerical Columns',fontsize=20)
for i in range(df.shape[1]):
  plt.subplot(10,3,i+1)
  f=plt.gca()
  f.set_title(df.columns.values[i],fontsize=16)

  val=np.size(df.iloc[:,i].unique())
  if val>=100:
    vals=100

  plt.hist(df.iloc[:,i],bins=vals,color='dodgerblue')
  plt.tight_layout(rect=[0,0.3,1,0.95])

  plt.xlabel(df.columns.values[i],fontsize=16)
  plt.ylabel('Frequency',fontsize=16)
plt.show()

```

```python
fraud=df['Class'].value_counts()[1]
normal=df['Class'].value_counts()[0]
ratio=(fraud/normal)*100
print('Fraudulent: {fraud},Non-Fraudaulent: {normal}'.format(fraud=fraud,normal=normal))
print('Percentage of Fraudulent: {ratio}'.format(ratio=ratio))
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=df['Class'].value_counts().index,y=df['Class'].value_counts().values)
plt.title('Fraudulent vs Non-Fraudulent Transactions')
plt.ylabel('Count')
plt.xlabel('Class')
plt.show()
```

```python
df.iloc[:,:-1].corrwith(df['Class']).plot(kind='bar',figsize=(20,7),
                                         color='lightcoral',
                                         fontsize=15,
                                         rot=45,
                                         grid=True)
plt.title('Correlation with Class',fontsize=20)
plt.show()
```

```python
# correlation matrix between the features
plt.figure(figsize=(10,8))
sns.set(font_scale=1.4)
sns.heatmap(df.corr(),
            linewidths=.5,
            cmap="RdBu_r",
            annot_kws={"size":10},
            cbar_kws={'shrink':.3}
           )
plt.xticks(rotation=45)
plt.title('Correlation Matrix', fontsize=25)
plt.show()
```

```python
#Training dataset into taining set and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
classifier = DecisionTreeClassifier(random_state=0,criterion='entropy')
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
plt.figure(figsize=(20,10))
tree.plot_tree(classifier, filled=True, feature_names=df.columns[:-1], class_names=['Non-Fraudulent', 'Fraudulent'])
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
knn_pred = knn.predict(X_test)
print(np.concatenate((knn_pred.reshape(len(knn_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print('Classification Report: \n\n',classification_report(y_test, y_pred))
plt.figure(figsize=(10,6))
sns.set(font_scale=1.4)
sns.heatmap(cm,
            linewidths=.5,
            annot=True,
            fmt='g',
            cmap="RdBu_r"
           )
plt.title('Confusion Matrix', fontsize=14)
plt.show()
```
